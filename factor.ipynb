```
def log_return(stock_ls):
  log_return_ls=[]
  for m in range(len(stock_ls)):
    stock = stock_ls[m]
    log_stock=np.log(stock)
    log_return1=log_stock[1:].copy()
    for i in range(len(log_return1)):
        log_return1[i]=log_stock[i+1]-log_stock[i]
    log_return_ls.append(log_return1)
  return log_return_ls

def mu(r_ls):
  mu1=[]
  for i in range(len(r_ls)):
    mu1.append(np.mean(r_ls[i]))
  return np.array(mu1).reshape(-1,1)

def cov(r_ls):
  df_temp={}
  for i in range(len(r_ls)):
    df_temp[index_list[i]]=r_ls[i]
  df = pd.DataFrame(df_temp)
  cov_mat=df.cov()  
  return cov_mat

def net_return(stock_ls):
  net_return_ls=[]
  for m in range(len(stock_ls)):
    stock = stock_ls[m]
    #log_stock=np.log(stock)
    net_return1=stock[1:].copy()
    for i in range(len(net_return1)):
        net_return1[i]=(stock[i+1]/stock[i])-1
    net_return_ls.append(net_return1)
  return net_return_ls

logreturn3=log_return(stock_list)
mu3=mu(logreturn3)
cov33=cov(logreturn3)
```
